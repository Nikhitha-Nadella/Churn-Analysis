```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('customer churn.csv')
df
```

```python
df.info()
```

#replacing blanks in total charges column as 0 as the tenure is 0 and there are no total charges 
#changed the datatype of the column to 'Float' from 'Object'

```python
df["TotalCharges"]=df["TotalCharges"].replace(" " ,"0")
df["TotalCharges"]=df["TotalCharges"].astype("float")
```

# coverted 0 and 1 values of senior citizen to "yes" or "no" for better understanding

```python
def conv(value):
    if value==1:
        return "yes"
    else:
        return "no"
df['SeniorCitizen']=df['SeniorCitizen'].apply(conv)
```

#Converetd the churn data into groupby as the pie chart leaves an error if it is plotted as it is because a string data cannot be convereted into a plot and percentages cannot be shown directly without grouing the data.

```python
plt.figure(figsize=(4,4))
gb=df.groupby('Churn').agg({'Churn':"count"})
plt.pie(gb['Churn'], labels=gb.index, autopct="%1.2f%%")
plt.title("Percentage Of Churned Customers", fontsize=10)
plt.show()
```

#numeric version

```python

# Step 1: Create contingency table
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Step 2: Plot
plt.figure(figsize=(4, 4))
bottom = [0] * len(counts)

for churn_status in counts.columns:
    values = counts[churn_status]
    bars = plt.bar(
        counts.index, 
        values, 
        bottom=bottom, 
        label=churn_status
    )

    # Step 3: Add labels
    for i, (bar, value) in enumerate(zip(bars, values)):
        height = bar.get_height()
        if height > 0:
            plt.text(
                bar.get_x() + bar.get_width() / 2, 
                bar.get_y() + height / 2, 
                str(value),
                ha='center', 
                va='center', 
                fontsize=10,
                color='white' if height > 15 else 'black'  # contrast
            )

    # Update bottom for stacking
    bottom = [b + v for b, v in zip(bottom, values)]

# Formatting
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.xlabel("Senior Citizen")
plt.ylabel("Count")
plt.title("Churn by Senior Citizen (Stacked Bar Chart)")
plt.legend(title="Churn")
plt.tight_layout()
plt.show()
```

#Percentage version

```python

# Step 1: Get raw counts and normalize to get percentages
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

# Step 2: Plot
plt.figure(figsize=(4, 4))
bottom = [0] * len(percentages)

for churn_status in percentages.columns:
    values = percentages[churn_status]
    bars = plt.bar(
        percentages.index, 
        values, 
        bottom=bottom, 
        label=churn_status
    )

    # Step 3: Add percentage labels
    for i, (bar, value) in enumerate(zip(bars, values)):
        if value > 3:  # Only label if segment is big enough to avoid clutter
            plt.text(
                bar.get_x() + bar.get_width() / 2, 
                bar.get_y() + value / 2, 
                f"{value:.1f}%", 
                ha='center', 
                va='center', 
                fontsize=10,
                color='white' if value > 15 else 'black'
            )

    # Update bottom for stacking
    bottom = [b + v for b, v in zip(bottom, values)]

# Formatting
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.title("Churn by Senior Citizen (100% Stacked Bar Chart)")
plt.legend(title="Churn")
plt.ylim(0, 100)
plt.show()
```

```python
# List of categorical columns
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Create a grid: 3 rows × 3 columns
fig, axes = plt.subplots(3, 3, figsize=(18, 12))
axes = axes.flatten()  # Flatten 2D array of axes to 1D for easy looping

# Loop through columns and plot countplot on each subplot
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i], hue="Churn")
    axes[i].set_title(f"Count Plot of {col}")
    axes[i].tick_params(axis='x', rotation=0)  # Rotate labels for readability

# Remove any unused subplots (if fewer than total grid slots)
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#Churn is higher among customers lacking services like OnlineSecurity, OnlineBackup, TechSupport, and DeviceProtection. Customers with fiber optic internet also show a notably higher churn rate. Overall, lack of value-added services seems to correlate with increased churn.
